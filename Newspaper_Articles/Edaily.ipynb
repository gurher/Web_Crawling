```python
# Import required modules

from bs4 import BeautifulSoup
import urllib.request
from urllib import parse

#Functions

def Edaily(search = 'None', save_path = 'c:',file_name = 'Edaily', page_num = 1):
  p1 = 'https://www.edaily.co.kr/search/news/?keyword='
  p2 =  parse.quote(search)
  p3 =  '&page='
  p4 = str(page_num)
  list_url = p1 + p2 + p3 + p4

  url = urllib.request.Request(list_url) 
  res = urllib.request.urlopen(url).read().decode("utf-8") 
  stuff = BeautifulSoup( res, "html.parser")
  temp = stuff.select(" #newsList > div > a > ul > li ")
  
  articles = ''
  websites = []
  for index,article in enumerate(temp):
    if index%2 == 1:
      websites.append(article.text)
      articles += article.text

  
  with open(save_path + '\\' + file_name + '.txt', 'w', encoding = 'utf8') as f:
    for x in websites:
        f.write(x)
        f.write('\n')
        f.write('\n')

  return articles
```
